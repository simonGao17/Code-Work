Creat Dataframe from CSV Data File

```python
df = pd.read_csv('EbayPcLaptopDataUnclean.csv')
df1 = pd.read_csv('EbayPcLaptopPriceData.csv')
```

Combine two Datasets
Merge with the EbayPcLaptopDataUnclean.csv using "item number"

```python
df = pd.merge(df,df1, on='Item Number', how ='left')
print(df)
```

According to the first 20 rows, we found that there are several instances of MISSING VALUES as indicated by the "NaN" value. There are also INCONSISTENT VALUE FORMATES entered into the "Screen Size", "Released Year", etc.

So now, let's handling missing data

```python
# Replace the field values with Fales or True based on MISSING VALUES
df.isna()
# check how many missing values in each column
missing = df.isna().sum()
print()
```

```python
# convert those numbers into percentages by dividing the counts by the number of records
missingpercentages = missing/len(df)*100
print(missingpercentages)
```

Considering that there will be no missing values in the dataset, and we can not add any additonal information to replace the "NaN". We should delete all missing values. But if we delete all missing values right now, there will be no any data in the dataset because some columns missed too much values. We should consider drop some columns first and then delete the missing value.

Let's drop the columns with high percentage of missing value. 
Ratings Count, Rating, Manufacturer Color, Country Region of Manufacture

```python
df = df.drop(columns='Rating')
df = df.drop(columns='Ratings Count')
df = df.drop(columns='Manufacturer Color')
df = df.drop(columns='Country Region Of Manufacture')
```

```python
# Then drop all missing values
df = df.dropna()
df
```

Thus, right now we have 139 rows * 20 columns with all values in the dataset. To confirm that we dropped all records missing values, let us re-run the counts on our new dataframe.

```python
# check if we still have how many missing value in each column
missing = df.isna().sum()
print(missing)
```
